```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("covid-19.csv")
```

```python
pd.set_option('display.max_rows', None)

df
```

```python
df[df['Country/Region'].isin(['US', 'Italy'])]
```

```python
x = df[df['Country/Region'].isin(['US', 'Italy', 'Germany', 'Spain', 'United Kingdom'])].groupby('Country/Region').sum().iloc[:, 2:]
```

```python
pd.set_option('display.max_columns', None)
x
```

```python
x.T.plot(legend='reverse', title="Number of confirmed cases over time")
```
